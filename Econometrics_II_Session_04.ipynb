## More on Dictionaries

### Adding, Changing, and Deleting Elements in Dictionaries

```python
players_numbers = dict(Muller = 25 , Neuer = 1)
players_numbers
```

```python
print(players_numbers["Muller"])
```

```python
grades = {
    "Robert": {"math": 9.5, "geography": 19},
    "Marcus": {"math": 15, "geography": 18.5}
}
print(grades)
```

```python
print(grades["Robert"]["geography"])
```

```python
players_numbers["Lewandowski"] = 5
```

```python
players_numbers
```

```python
# First, create the nested dictionary for Penny
grades["Penny"] = {}

# Then, assign values to the keys in the nested dictionary
grades["Penny"]["math"] = 0
grades["Penny"]["geography"] = 0.25

print(grades)
```

```python
del grades["Marcus"]
grades
```

### Performing mathematics with dicts

#### addition using sum()

```python
data_age = {"Sheldon" : 37, "Leonard" : 38, "Howard" : 35}
```

```python
total_sum = sum(data_age.values()) #keys : values
print(total_sum)
```

#### multiplication using prod()

```python
import math
age_product = math.prod(data_age.values())
print(age_product)
```

```python
37 * 38 * 35
```

#### Maximum and Minimum values

```python
min(data_age.values())
```

```python
max(data_age.values())
```

#### Calculating Averages

```python
age_average = sum(data_age.values()) / len(data_age)
print(age_average)
```
