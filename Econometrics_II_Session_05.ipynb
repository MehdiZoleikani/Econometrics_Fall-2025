### Functions

Understanding Function Definitions:

When defining a function in Python, you can specify one or more parameters inside parentheses — these act as placeholders for the values that will be passed to the function when it’s called.  
If no parameters are provided, the function simply executes its internal code block.  
Every function definition must end with a colon (`:`), and the indented lines that follow represent the body of the function — that is, the set of instructions that the function will perform.

#### What is the difference between using a local variable and not using a local variable in a function?

```python
def area(length, width):
    x = length * width #local variable
    return x
```

```python
area(3, 2)
```

In this example, x is stored in memory and further calculations can be done on it, but with pure return statement, the result is temporary stored. For example:

```python
def area(length, width):
    x = length * width
    # Further calculations
    perimeter = 2 * (length + width)
    return x, perimeter
```

```python
area(3, 2)
```

**30 is x or area and 22 is the perimeter**

In the above example, only return is being used and you can clearly see that only one of the line is printed, which is the area and perimeter is left out.

#### Adding description

##### Can be added within the function:

```python
def area(length, width):
    x = length * width
    perimeter = 2 * (length + width)
    print(f"Area is {x}.")
    print(f"Perimeter is {perimeter}.")
    return x, perimeter
```

```python
area(3, 2)
```

##### Can be added outside of the function:

```python
def area(length, width):
    x = length * width 
    perimeter = 2 * (length + width)
    return x, perimeter

area_value, perimeter_value = area(5, 3) #multiple variable assignment

print(f"Area is {area_value}")
print(f"Perimeter is {perimeter_value}")
```

### Control Flow Structures

#### if-elif-else

```python
list2 = [-1, -2, 0, 1, 2]
```

```python
for i in list2:
    if i > 0:
        print(i)
```

```python
for i in list2:
    if i > 0: #main condition
        print(f"The positive number: {i}")
    elif i == 0: #alternative condition (else if)
        print(f"Neither positive nor negative number: {i}")
    else: #for all other cases
        print(f"The negative number: {i}")
```

```python
for i in list2:
    if i > 0:
        print(i, "This number is greater than 0")
    else:
        print(i, "This number is smaller than or equal to 0")
```
