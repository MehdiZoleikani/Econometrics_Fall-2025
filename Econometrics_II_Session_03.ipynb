#### For gamers: Health Points updates each time you run this code.

```python
#HP: Player's Health points at the start of the game.
HP = 100
print("Player's Health points at the start of the game:", HP)
```

```python
#Player gets shot in the right arm!
HP -= 5  
print("Hit in the right arm! HP:", HP)
```

```python
#Player gets shot in the neck!
HP -= 90
print(f"Hit in the neck! HP: {HP}") #f-string
```

```python
#Player uses a medkit!
HP += 30  
print("Used a medkit! HP:", HP)
```

```python
print("Game over!" if HP <= 0 else "Keep playing!")
```
