## Solving a system of equations using cramer's rule

\begin{cases}
3x + 2y = 10 \\
5x - y = 7
\end{cases}

```python
A = np.array([[3, 2], [5, -1]])
A
```

```python
B = np.array([10, 7])
B
```

The determinant measures how a matrix scales space and indicates whether the matrix is invertible (det ≠ 0).

```python
determinant_A = np.linalg.det(A)
determinant_A
```

```python
A_x = A.copy()
B_x = A.copy()
```

for finding x, we want to replace the first column of matrix A with matrix B.

```python
A_x
```

```python
A_x[:, 0]
```

```python
A_x[:, 0] = B
print(A_x)
```

```python
np.linalg.det(A_x)
```

```python
B_x[:, 1]
```

```python
B_x[:, 1] = B
print(B_x)
```

np.linalg.det(B_x)

### Finding x

```python
x = np.linalg.det(A_x) / determinant_A
print(x)
```

### Finding y

```python
y = np.linalg.det(B_x) / determinant_A
print(y)
```
